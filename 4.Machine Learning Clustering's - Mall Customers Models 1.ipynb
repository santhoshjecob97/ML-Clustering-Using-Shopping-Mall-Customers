```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import (KMeans, AffinityPropagation, MeanShift, SpectralClustering,
                             AgglomerativeClustering, DBSCAN, OPTICS, Birch)
from sklearn.metrics import silhouette_score
```

```python
class ClusterModels:
    def __init__(self, X):
        self.X = X

    def evaluate(self, labels):
        if len(set(labels)) > 1:
            score = silhouette_score(self.X, labels)
            print(f"Silhouette Score: {score:.2f}")
        else:
            print("Only one cluster found. Silhouette score is not applicable.")

    def plot_clusters(self, labels, title):
        plt.figure(figsize=(6, 5))
        plt.scatter(self.X[:, 0], self.X[:, 1], c=labels, cmap='viridis', s=50)
        plt.title(title)
        plt.xlabel("Annual Income (k$)")
        plt.ylabel("Spending Score (1-100)")
        plt.grid(True)
        plt.show()

    def run_kmeans(self, n_clusters=5):
        model = KMeans(n_clusters=n_clusters, random_state=5)
        labels = model.fit_predict(self.X)
        print("KMeans Clustering")
        self.evaluate(labels)
        self.plot_clusters(labels, "KMeans")
        return labels
    
    def run_affinity(self):
        model = AffinityPropagation()  # FIXED
        labels = model.fit_predict(self.X)
        print("Affinity Propagation")
        self.evaluate(labels)
        self.plot_clusters(labels, "Affinity Propagation")
        return labels
    
    def run_mean_shift(self):
        model = MeanShift()
        labels = model.fit_predict(self.X)
        print("Mean Shift Clustering")
        self.evaluate(labels)
        self.plot_clusters(labels, "Mean Shift")
        return labels

    def run_spectral(self, n_clusters=5):
        model = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=5)
        labels = model.fit_predict(self.X)
        print("Spectral Clustering")
        self.evaluate(labels)
        self.plot_clusters(labels, "Spectral Clustering")
        return labels

    def run_agglomerative(self, n_clusters=5):
        model = AgglomerativeClustering(n_clusters=n_clusters)
        labels = model.fit_predict(self.X)
        print("Agglomerative Clustering")
        self.evaluate(labels)
        self.plot_clusters(labels, "Agglomerative Clustering")
        return labels

    def run_dbscan(self, eps=6, min_samples=5):
        model = DBSCAN(eps=eps, min_samples=min_samples)
        labels = model.fit_predict(self.X)
        print("DBSCAN Clustering")
        self.evaluate(labels)
        self.plot_clusters(labels, "DBSCAN")
        return labels

    def run_optics(self, min_samples=5):
        model = OPTICS(min_samples=min_samples)
        labels = model.fit_predict(self.X)
        print("OPTICS Clustering")
        self.evaluate(labels)
        self.plot_clusters(labels, "OPTICS")
        return labels

    def run_birch(self, n_clusters=5):
        model = Birch(n_clusters=n_clusters)
        labels = model.fit_predict(self.X)
        print("BIRCH Clustering")
        self.evaluate(labels)
        self.plot_clusters(labels, "BIRCH")
        return labels


# === Run all clustering models ===
clusterer = ClusterModels(X)

clusterer.run_kmeans()
clusterer.run_affinity()
clusterer.run_mean_shift()
clusterer.run_spectral()
clusterer.run_agglomerative()
clusterer.run_dbscan()
clusterer.run_optics()
clusterer.run_birch()
```
