```python
# Shopping Mall Clustering - All Major Clustering Models

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python

# Load dataset
df = pd.read_csv("Mall_Customers.csv")
df['Genre'] = LabelEncoder().fit_transform(df['Genre'])

# Feature selection and scaling
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Genre']
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dimensionality reduction for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Define clustering methods
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import (KMeans, AffinityPropagation, MeanShift, SpectralClustering,
                             AgglomerativeClustering, DBSCAN, OPTICS, Birch)
from sklearn.mixture import GaussianMixture
clustering_models = {
    "KMeans": KMeans(n_clusters=5, random_state=42),
    "Agglomerative": AgglomerativeClustering(n_clusters=5),
    "AffinityPropagation": AffinityPropagation(damping=0.9),
    "MeanShift": MeanShift(),
    "Spectral": SpectralClustering(n_clusters=5, affinity='nearest_neighbors', random_state=42),
    "DBSCAN": DBSCAN(eps=0.5, min_samples=5),
    "OPTICS": OPTICS(min_samples=5),
    "GaussianMixture": GaussianMixture(n_components=5, random_state=42),
    "Birch": Birch(n_clusters=5)
}

results = []

# Plotting function
def plot_clusters(model_name, labels):
    plt.figure(figsize=(6, 5))
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='rainbow', s=25)
    plt.title(f"{model_name} Clustering")
    plt.xlabel("PCA 1")
    plt.ylabel("PCA 2")
    plt.show()

# Run and evaluate each model
for name, model in clustering_models.items():
    try:
        print(f"Running: {name}")
        start_time = time.time()

        if name == "GaussianMixture":
            model.fit(X_scaled)
            labels = model.predict(X_scaled)
        else:
            labels = model.fit_predict(X_scaled)

        runtime = time.time() - start_time
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

        score = silhouette_score(X_scaled, labels) if n_clusters > 1 else -1
        results.append((name, n_clusters, round(score, 3), round(runtime, 3)))

        plot_clusters(name, labels)

    except Exception as e:
        print(f"Model {name} failed: {e}")
        results.append((name, "Error", "Error", "Error"))

# Summary table

from sklearn.metrics import silhouette_score
summary = pd.DataFrame(results, columns=["Model", "Clusters", "Silhouette Score", "Runtime (s)"])
print("\nModel Comparison Summary:")
print(summary)
sns.barplot(data=summary[summary['Silhouette Score'] != "Error"], x="Silhouette Score", y="Model")
plt.title("Silhouette Score Comparison")
plt.show()
```
