```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python

# Load dataset
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values  # Using Income and Spending Score

# Scale the data
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

# Reduce to 2D for plotting
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

```

```python
# Store results
model_names = []
cluster_counts = []
silhouette_scores = []

```

```python
# Clustering models
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, MeanShift, SpectralClustering, AffinityPropagation, OPTICS, Birch
from sklearn.mixture import GaussianMixture

def safe_silhouette(X, labels):
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    return silhouette_score(X, labels) if n_clusters > 1 else -1

# Run each model

# KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
model_names.append("KMeans")
cluster_counts.append(len(set(y_kmeans)))
silhouette_scores.append(safe_silhouette(X_scaled, y_kmeans))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans)
plt.title("KMeans")
plt.show()

# Agglomerative
agg = AgglomerativeClustering(n_clusters=5)
y_agg = agg.fit_predict(X_scaled)
model_names.append("Agglomerative")
cluster_counts.append(len(set(y_agg)))
silhouette_scores.append(safe_silhouette(X_scaled, y_agg))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_agg)
plt.title("Agglomerative")
plt.show()

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_db = dbscan.fit_predict(X_scaled)
model_names.append("DBSCAN")
cluster_counts.append(len(set(y_db)) - (1 if -1 in y_db else 0))
silhouette_scores.append(safe_silhouette(X_scaled, y_db))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_db)
plt.title("DBSCAN")
plt.show()

# MeanShift
ms = MeanShift()
y_ms = ms.fit_predict(X_scaled)
model_names.append("MeanShift")
cluster_counts.append(len(set(y_ms)))
silhouette_scores.append(safe_silhouette(X_scaled, y_ms))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_ms)
plt.title("MeanShift")
plt.show()

# Spectral Clustering
scm = SpectralClustering(n_clusters=5, affinity='nearest_neighbors')
y_spectral = scm.fit_predict(X_scaled)
model_names.append("Spectral")
cluster_counts.append(len(set(y_spectral)))
silhouette_scores.append(safe_silhouette(X_scaled, y_spectral))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_spectral)
plt.title("Spectral Clustering")
plt.show()

# Affinity Propagation
ap = AffinityPropagation(damping=0.9)
y_ap = ap.fit_predict(X_scaled)
model_names.append("AffinityPropagation")
cluster_counts.append(len(set(y_ap)))
silhouette_scores.append(safe_silhouette(X_scaled, y_ap))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_ap)
plt.title("Affinity Propagation")
plt.show()

# OPTICS
opt = OPTICS(min_samples=5)
y_opt = opt.fit_predict(X_scaled)
model_names.append("OPTICS")
cluster_counts.append(len(set(y_opt)) - (1 if -1 in y_opt else 0))
silhouette_scores.append(safe_silhouette(X_scaled, y_opt))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_opt)
plt.title("OPTICS")
plt.show()

# Birch
birch = Birch(n_clusters=5)
y_birch = birch.fit_predict(X_scaled)
model_names.append("Birch")
cluster_counts.append(len(set(y_birch)))
silhouette_scores.append(safe_silhouette(X_scaled, y_birch))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_birch)
plt.title("Birch")
plt.show()

# Gaussian Mixture
gm = GaussianMixture(n_components=5, random_state=42)
y_gm = gm.fit_predict(X_scaled)
model_names.append("GaussianMixture")
cluster_counts.append(len(set(y_gm)))
silhouette_scores.append(safe_silhouette(X_scaled, y_gm))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_gm)
plt.title("Gaussian Mixture")
plt.show()


# Result Table
result_df = pd.DataFrame({
    "Model": model_names,
    "Clusters": cluster_counts,
    "Silhouette Score": silhouette_scores
})


```

```python
# Find best model
best_index = result_df["Silhouette Score"].idxmax()
best_model = result_df.loc[best_index, "Model"]
print("\nAll Model Results:")
print(result_df)
print(f"\n✅ Best model based on silhouette score: {best_model}")

# Optional: Plot the best model with original data
# Here we show Agglomerative as supervised example
label = y_agg  # You can also use the best model label here

```

```python
from sklearn.base import BaseEstimator, ClusterMixin
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Your custom clustering model (same as before)
class QuadCrossCluster(BaseEstimator, ClusterMixin):
    def __init__(self, angle=0):
        self.angle = angle
        self.centroid_ = None
        self.labels_ = None
        
    def fit(self, X):
        self.centroid_ = np.mean(X, axis=0)
        centered = X - self.centroid_
        
        if self.angle != 0:
            theta = np.radians(self.angle)
            rot_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                                 [np.sin(theta), np.cos(theta)]])
            centered = np.dot(centered, rot_matrix)
        
        self.labels_ = np.zeros(X.shape[0], dtype=int)
        self.labels_[(centered[:, 0] >= 0) & (centered[:, 1] >= 0)] = 0
        self.labels_[(centered[:, 0] < 0) & (centered[:, 1] >= 0)] = 1
        self.labels_[(centered[:, 0] < 0) & (centered[:, 1] < 0)] = 2
        self.labels_[(centered[:, 0] >= 0) & (centered[:, 1] < 0)] = 3
        
        return self
    
    def fit_predict(self, X):
        return self.fit(X).labels_

# Initialize lists for storing results
model_names = []
cluster_counts = []
silhouette_scores = []

# Safe silhouette score function
def safe_silhouette(X, labels):
    from sklearn.metrics import silhouette_score
    if len(set(labels)) < 2:
        return -1  # Return invalid score if only one cluster
    return silhouette_score(X, labels)

# Assuming you have X_scaled and X_pca already defined
# X_scaled = your scaled feature matrix
# X_pca = your PCA-transformed data (2D)

# Create and fit your custom model
qcc = QuadCrossCluster(angle=45)  # Adjust angle as needed
y_qcc = qcc.fit_predict(X_scaled)

# Store results
model_names.append("Quad-Cross")
cluster_counts.append(len(set(y_qcc)))
silhouette_scores.append(safe_silhouette(X_scaled, y_qcc))

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_qcc, cmap='viridis')
plt.title("Quad-Cross Clustering")
plt.colorbar(label='Cluster')
plt.grid(True, alpha=0.3)
plt.show()

# Print metrics
print(f"Quad-Cross Clustering Results:")
print(f"Number of clusters: {len(set(y_qcc))}")
print(f"Silhouette Score: {silhouette_scores[-1]:.4f}")
```

```python
from sklearn.base import BaseEstimator, ClusterMixin
import numpy as np
import matplotlib.pyplot as plt

class RhombusCluster(BaseEstimator, ClusterMixin):
    def __init__(self, angle=45):
        self.angle = angle  # Controls rhombus orientation
        self.centroid_ = None
        self.labels_ = None
        
    def fit(self, X):
        self.centroid_ = np.mean(X, axis=0)
        centered = X - self.centroid_
        
        # Rotate data to align with rhombus axes
        theta = np.radians(self.angle)
        rot_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                             [np.sin(theta), np.cos(theta)]])
        rotated = np.dot(centered, rot_matrix)
        
        # Assign clusters based on diamond-shaped boundaries
        self.labels_ = np.zeros(X.shape[0], dtype=int)
        
        # Calculate diamond boundaries (sum of absolute values)
        boundary = np.abs(rotated[:,0]) + np.abs(rotated[:,1])
        
        # Create 4 clusters in diamond pattern
        self.labels_[(rotated[:,0] >= 0) & (rotated[:,1] >= 0)] = 0  # Top-right
        self.labels_[(rotated[:,0] < 0) & (rotated[:,1] >= 0)] = 1    # Top-left
        self.labels_[(rotated[:,0] < 0) & (rotated[:,1] < 0)] = 2     # Bottom-left
        self.labels_[(rotated[:,0] >= 0) & (rotated[:,1] < 0)] = 3    # Bottom-right
        
        return self
    
    def fit_predict(self, X):
        return self.fit(X).labels_

# Usage example (same format as before)
rhombus = RhombusCluster(angle=45)
y_rhombus = rhombus.fit_predict(X_scaled)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_rhombus, cmap='viridis')
plt.title("Rhombus Clustering")

# Draw diamond-shaped boundaries
centroid = np.mean(X_pca, axis=0)
for angle in [45, 135, 225, 315]:
    x = centroid[0] + 3 * np.cos(np.radians(angle))
    y = centroid[1] + 3 * np.sin(np.radians(angle))
    plt.plot([centroid[0], x], [centroid[1], y], 'r--', alpha=0.5)

plt.colorbar(label='Cluster')
plt.show()

# Print metrics
print(f"Rhombus Clustering  Results:")
print(f"Number of clusters: {len(set(y_qcc))}")
print(f"Silhouette Score: {silhouette_scores[-1]:.4f}")
```

```python

from sklearn.base import BaseEstimator, ClusterMixin
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

class OctagonCluster(BaseEstimator, ClusterMixin):
    def __init__(self, angle=30):
        self.angle = angle  # Rotation angle in degrees
        self.centroid_ = None
        self.labels_ = None
        
    def fit(self, X):
        self.centroid_ = np.mean(X, axis=0)
        centered = X - self.centroid_
        
        # Apply rotation
        theta = np.radians(self.angle)
        rot_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                             [np.sin(theta), np.cos(theta)]])
        rotated = np.dot(centered, rot_matrix)
        
        # Calculate angles for octant assignment
        angles = np.degrees(np.arctan2(rotated[:,1], rotated[:,0])) % 360
        self.labels_ = np.floor(angles / 45).astype(int)  # 8 octants
        
        return self
    
    def fit_predict(self, X):
        return self.fit(X).labels_

# Create and fit the model
octagon = OctagonCluster(angle=22.5)  # 22.5° rotation aligns octants with axes
y_octagon = octagon.fit_predict(X_scaled)

# Calculate silhouette score
sil_score = silhouette_score(X_scaled, y_octagon)
print(f"Octagon Clustering Silhouette Score: {sil_score:.4f}")

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_octagon, cmap='tab10', s=30)
plt.title(f"Octagon Clustering (Silhouette: {sil_score:.3f})")

# Draw octagon boundaries
centroid = np.mean(X_pca, axis=0)
for i in range(8):
    angle = np.radians(i * 45 + 22.5)
    x = centroid[0] + 3 * np.cos(angle)
    y = centroid[1] + 3 * np.sin(angle)
    plt.plot([centroid[0], x], [centroid[1], y], 'k--', alpha=0.3)

plt.colorbar(label='Octant')
plt.grid(True, alpha=0.2)
plt.show()

# Print metrics
print(f"Octogan Clustering  Results:")
print(f"Number of clusters: {len(set(y_qcc))}")
print(f"Silhouette Score: {silhouette_scores[-1]:.4f}")
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt

def auto_gmm(X, max_components=10):
    best_score = -1
    best_gmm = None
    
    for n in range(2, max_components+1):
        gmm = GaussianMixture(n_components=n, random_state=42)
        labels = gmm.fit_predict(X)
        score = silhouette_score(X, labels)
        
        if score > best_score:
            best_score = score
            best_gmm = gmm
            
    return best_gmm.fit_predict(X)

# Assuming X_scaled is your scaled data and X_pca is 2D PCA-transformed data
labels_gmm = auto_gmm(X_scaled)
sil_score = silhouette_score(X_scaled, labels_gmm)

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_gmm, cmap='tab10', s=30)
plt.title(f"Optimized GMM Clustering (Silhouette: {sil_score:.3f})")
plt.colorbar(label='Cluster')
plt.grid(True, alpha=0.2)
plt.show()

# Print metrics
print(f"Clustering Results:")
print(f"Number of clusters: {len(np.unique(labels_gmm))}")
print(f"Silhouette Score: {sil_score:.4f}")
```
